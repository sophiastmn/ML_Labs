```python
#################################### Problem #1 ###########################################
#imports
from sklearn.svm import SVC;
import matplotlib.pyplot as plot;
from sklearn.datasets import make_blobs;
from sklearn.inspection import DecisionBoundaryDisplay;

#global variables
seed = 1880333;

def main():
  #make the plot
  x1, y1 = make_blobs(n_samples = 100, n_features = 2, centers = [[2, 1], [1, 4.5]], cluster_std = 0.5, random_state=seed);
  plot.scatter(x1[:, 0], x1[:, 1], c=y1, s=50, cmap="Accent");
  plot.show();

  #do the SVM
  model = SVC(kernel="linear", C=1.0, random_state=seed);
  model.fit(x1, y1);
  disp = DecisionBoundaryDisplay.from_estimator(model, x1, response_method="predict",
                                                cmap="Accent", alpha=0.3);
  plot.scatter(x1[:, 0], x1[:, 1], c=y1, s=50, cmap="Accent");
  plot.show();
  plot.cla();


  #experiment with c
  for i in range(1, 21):
    print(f"C value: {i*0.1}");
    model = SVC(kernel="linear", C=i*0.1, random_state=seed);
    model.fit(x1, y1);
    disp = DecisionBoundaryDisplay.from_estimator(model, x1, response_method="predict",
                                                cmap="Accent", alpha=0.3);
    plot.scatter(x1[:, 0], x1[:, 1], c=y1, s=50, cmap="Accent");
    plot.show();
    plot.cla();


if __name__ == "__main__":
  main();

```

```python
#################################### Problem #2 ###########################################
#imports
from sklearn.svm import SVC;
import matplotlib.pyplot as plot;
from sklearn.datasets import make_blobs;
from sklearn.inspection import DecisionBoundaryDisplay;
from sklearn.datasets import load_iris;

#global variables
seed = 1880333;

def main():
  #get the iris data
  x1_full = load_iris().data;
  x1 = x1_full[:, :2];
  y1 = load_iris().target[range(0, 150)];

  #do the SVM
  model = SVC(kernel='linear', C = 1.0, decision_function_shape='ovo', random_state=seed);
  model.fit(x1, y1);
  disp = DecisionBoundaryDisplay.from_estimator(model, x1, response_method="predict",
                                                cmap="Accent", alpha=0.3);
  plot.scatter(x1[:, 0], x1[:, 1], c=y1, s=50, cmap="Accent");
  plot.show();
  plot.cla();


if __name__ == "__main__":
  main();
```
